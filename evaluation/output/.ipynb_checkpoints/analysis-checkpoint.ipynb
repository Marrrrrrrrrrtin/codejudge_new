```python
import pandas as pd
import os
import json

# Step 1: Load all JSONs from your folder
folder_path = "./humaneval/python-small-test/"
json_files = [f for f in os.listdir(folder_path) if f.endswith(".json")]

all_entries = []

for file in json_files:
    file_path = os.path.join(folder_path, file)
    with open(file_path, "r") as f:
        content = json.load(f)
    for entry in content.get("data", []):
        entry["source_file"] = file
        all_entries.append(entry)

df = pd.DataFrame(all_entries)

# Step 2: Flatten `code_gpt_score` subfields
score_df = pd.json_normalize(df["code_gpt_score"])

# Step 3: Merge back to the original dataframe
df = df.drop(columns=["code_gpt_score"]).join(score_df)

# Optional: rename columns if needed
df = df.rename(columns={
    "code_gpt_score": "gpt_score",
    "comparison": "gpt_comparison",
    "parsed_comparison": "gpt_parsed_comparison"
})


df.head()
```

```python
negative_truth = df[(df['gpt_score'] < 1) & (df['pass'])]
negative_truth
```

```python
positive_false = df[(df['gpt_score'] > 0) & (~df['pass'])]
positive_false
```

```python
import json
from pygments import highlight
from pygments.lexers import PythonLexer
from pygments.formatters import TerminalFormatter

def pretty_print_code_json(row):
    # Extract and format each code field
    formatted = {
        "pass": row["pass"],
        "question_id": row["question_id"],
        "program": highlight(
            row["program"].strip(), 
            PythonLexer(), 
            TerminalFormatter()
        ),
        "canonical_solution": highlight(
            row["canonical_solution"].strip(), 
            PythonLexer(), 
            TerminalFormatter()
        ),
        "code_gpt_score": row["code_gpt_score"]  # Leave as-is
    }
    
    print(json.dumps(formatted, indent=2))

# Example usage:
your_row = {
    "pass": False,
    "program": "def closest_integer(value):\n\n    return round(float(value))\n\n",
    # ... (your full JSON data)
}

pretty_print_code_json(your_row)
```
